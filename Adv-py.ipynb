```python
def find_duplicates(lst):
    seen = set()
    duplicates = set()

    for item in lst:
        if item in seen:
            duplicates.add(item)
        else:
            seen.add(item)

    return list(duplicates)

# Example
my_list = [1, 2, 3, 4, 2, 5, 1, 6]
print(find_duplicates(my_list))
```

```python
# Library Management System

library = []

def add_book():
    book_id = input("Enter Book ID: ")
    title = input("Enter Book Title: ")
    author = input("Enter Author Name: ")
    book = {
        "id": book_id,
        "title": title,
        "author": author,
        "issued": False
    }
    library.append(book)
    print("Book added successfully!\n")

def display_books():
    if not library:
        print("No books available.\n")
        return
    for book in library:
        status = "Issued" if book["issued"] else "Available"
        print(f'ID: {book["id"]}, Title: {book["title"]}, Author: {book["author"]}, Status: {status}')
    print()

def issue_book():
    book_id = input("Enter Book ID to issue: ")
    for book in library:
        if book["id"] == book_id:
            if not book["issued"]:
                book["issued"] = True
                print("Book issued successfully!\n")
            else:
                print("Book already issued!\n")
            return
    print("Book not found!\n")

def return_book():
    book_id = input("Enter Book ID to return: ")
    for book in library:
        if book["id"] == book_id:
            if book["issued"]:
                book["issued"] = False
                print("Book returned successfully!\n")
            else:
                print("Book was not issued!\n")
            return
    print("Book not found!\n")

# Main Menu
while True:
    print("📚 Library Management System")
    print("1. Add Book")
    print("2. Display Books")
    print("3. Issue Book")
    print("4. Return Book")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_book()
    elif choice == "2":
        display_books()
    elif choice == "3":
        issue_book()
    elif choice == "4":
        return_book()
    elif choice == "5":
        print("Thank you for using Library Management System!")
        break
    else:
        print("Invalid choice! Try again.\n")
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)
```

```python
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count
```

```python
def find_largest(lst):
    largest = lst[0]
    for num in lst:
        if num > largest:
            largest = num
    return largest
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

```python
def sum_of_digits(n):
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total
```

```python
def fibonacci(n):
    fib = []
    a, b = 0, 1
    for _ in range(n):
        fib.append(a)
        a, b = b, a + b
    return fib
```

```python
def remove_duplicates(lst):
    seen = set()
    result = []
    for item in lst:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return result
```

```python
def string_length(s):
    count = 0
    for _ in s:
        count += 1
    return count
```

```python
from datetime import datetime

# Parking data
slots = []
vehicles = {}

RATE_PER_HOUR = 20


def initialize_parking():
    total_slots = int(input("Enter total number of parking slots: "))

    for i in range(1, total_slots + 1):
        slots.append({
            "slot_id": i,
            "occupied": False,
            "vehicle_number": None,
            "entry_time": None
        })


def park_vehicle():
    vehicle_number = input("Enter vehicle number: ")
    vehicle_type = input("Enter vehicle type (Car/Bike/Truck): ")

    for slot in slots:
        if not slot["occupied"]:
            slot["occupied"] = True
            slot["vehicle_number"] = vehicle_number
            slot["entry_time"] = datetime.now()

            vehicles[vehicle_number] = slot

            print(f"\nVehicle parked in Slot {slot['slot_id']}")
            return

    print("Parking Full!")


def exit_vehicle():
    vehicle_number = input("Enter vehicle number: ")

    if vehicle_number not in vehicles:
        print("Vehicle not found!")
        return

    slot = vehicles[vehicle_number]
    exit_time = datetime.now()

    entry_time = slot["entry_time"]
    duration = (exit_time - entry_time).total_seconds() / 3600

    if duration < 1:
        duration = 1

    fee = duration * RATE_PER_HOUR

    print("\nVehicle Exit Successful")
    print("Slot Number:", slot["slot_id"])
    print("Parking Fee: ₹", round(fee, 2))

    slot["occupied"] = False
    slot["vehicle_number"] = None
    slot["entry_time"] = None

    del vehicles[vehicle_number]


def show_slots():
    print("\nParking Slot Status:")

    for slot in slots:
        if slot["occupied"]:
            print(f"Slot {slot['slot_id']} → Occupied")
        else:
            print(f"Slot {slot['slot_id']} → Free")


def menu():
    while True:
        print("\n--- Parking Lot Management ---")
        print("1. Park Vehicle")
        print("2. Exit Vehicle")
        print("3. Show Parking Slots")
        print("4. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            park_vehicle()

        elif choice == "2":
            exit_vehicle()

        elif choice == "3":
            show_slots()

        elif choice == "4":
            print("System Closed")
            break

        else:
            print("Invalid choice")


# Start program
initialize_parking()
menu()
```

```python
#freelanser
class Freelancer:
    def __init__(self, name, skill):
        self.name = name
        self.skill = skill
        self.balance = 0

class Client:
    def __init__(self, name):
        self.name = name

class Project:
    def __init__(self, title, client, budget):
        self.title = title
        self.client = client
        self.budget = budget
        self.freelancer = None
        self.completed = False


freelancers = []
clients = []
projects = []


def register_freelancer():
    name = input("Freelancer Name: ")
    skill = input("Skill: ")
    freelancers.append(Freelancer(name, skill))
    print("Freelancer registered successfully!")


def register_client():
    name = input("Client Name: ")
    clients.append(Client(name))
    print("Client registered successfully!")


def create_project():
    title = input("Project Title: ")
    client_name = input("Client Name: ")
    budget = float(input("Budget: "))

    client = next((c for c in clients if c.name == client_name), None)

    if client:
        projects.append(Project(title, client, budget))
        print("Project created successfully!")
    else:
        print("Client not found.")


def assign_freelancer():
    title = input("Project Title: ")
    freelancer_name = input("Freelancer Name: ")

    project = next((p for p in projects if p.title == title), None)
    freelancer = next((f for f in freelancers if f.name == freelancer_name), None)

    if project and freelancer:
        project.freelancer = freelancer
        print("Freelancer assigned successfully!")
    else:
        print("Project or Freelancer not found.")


def complete_project():
    title = input("Project Title: ")

    project = next((p for p in projects if p.title == title), None)

    if project and project.freelancer:
        project.completed = True
        project.freelancer.balance += project.budget
        print("Project completed and payment processed!")
    else:
        print("Project not found or freelancer not assigned.")


def view_data():
    print("\nFreelancers:")
    for f in freelancers:
        print(f.name, "-", f.skill, "- Balance:", f.balance)

    print("\nClients:")
    for c in clients:
        print(c.name)

    print("\nProjects:")
    for p in projects:
        freelancer = p.freelancer.name if p.freelancer else "Not Assigned"
        status = "Completed" if p.completed else "Pending"
        print(p.title, "| Client:", p.client.name, "| Freelancer:", freelancer, "| Status:", status)


while True:
    print("\n--- Freelancer Marketplace ---")
    print("1. Register Freelancer")
    print("2. Register Client")
    print("3. Create Project")
    print("4. Assign Freelancer")
    print("5. Complete Project / Process Payment")
    print("6. View Data")
    print("7. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        register_freelancer()
    elif choice == "2":
        register_client()
    elif choice == "3":
        create_project()
    elif choice == "4":
        assign_freelancer()
    elif choice == "5":
        complete_project()
    elif choice == "6":
        view_data()
    elif choice == "7":
        break
    else:
        print("Invalid choice")
```

```python
#ware house
 class Item:
    def __init__(self, name, quantity):
        self.name = name
        self.quantity = quantity
        self.sold_history = []

inventory = {}


def add_item():
    name = input("Enter item name: ")
    qty = int(input("Enter quantity: "))

    if name in inventory:
        inventory[name].quantity += qty
    else:
        inventory[name] = Item(name, qty)

    print("Item added to warehouse.")


def move_goods_in():
    name = input("Item name: ")
    qty = int(input("Quantity received: "))

    if name in inventory:
        inventory[name].quantity += qty
        print("Goods added to inventory.")
    else:
        print("Item not found.")


def move_goods_out():
    name = input("Item name: ")
    qty = int(input("Quantity shipped: "))

    if name in inventory:
        if inventory[name].quantity >= qty:
            inventory[name].quantity -= qty
            inventory[name].sold_history.append(qty)
            print("Goods shipped successfully.")
        else:
            print("Not enough stock.")
    else:
        print("Item not found.")


def inventory_report():
    print("\n--- Inventory Report ---")
    for item in inventory.values():
        print(f"Item: {item.name} | Stock: {item.quantity}")


def forecast_demand():
    print("\n--- Demand Forecast ---")
    for item in inventory.values():
        if item.sold_history:
            avg = sum(item.sold_history) / len(item.sold_history)
            print(f"{item.name} predicted demand: {round(avg)} units")
        else:
            print(f"{item.name} not enough sales data")


while True:
    print("\n===== Warehouse Automation System =====")
    print("1. Add Item")
    print("2. Goods Incoming")
    print("3. Goods Outgoing")
    print("4. Inventory Report")
    print("5. Forecast Demand")
    print("6. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_item()

    elif choice == "2":
        move_goods_in()

    elif choice == "3":
        move_goods_out()

    elif choice == "4":
        inventory_report()

    elif choice == "5":
        forecast_demand()

    elif choice == "6":
        break

    else:
        print("Invalid choice")
```

```python
#crm
class Customer:
    def __init__(self, name, email, phone):
        self.name = name
        self.email = email
        self.phone = phone
        self.logs = []
        self.stage = "Lead"

customers = {}


def add_customer():
    name = input("Customer Name: ")
    email = input("Email: ")
    phone = input("Phone: ")

    customers[name] = Customer(name, email, phone)
    print("Customer added successfully.")


def view_customers():
    if not customers:
        print("No customers found.")
        return

    for c in customers.values():
        print(f"\nName: {c.name}")
        print(f"Email: {c.email}")
        print(f"Phone: {c.phone}")
        print(f"Sales Stage: {c.stage}")


def add_log():
    name = input("Customer Name: ")

    if name in customers:
        log = input("Enter communication note: ")
        customers[name].logs.append(log)
        print("Log added.")
    else:
        print("Customer not found.")


def view_logs():
    name = input("Customer Name: ")

    if name in customers:
        print("\nCommunication Logs:")
        for log in customers[name].logs:
            print("-", log)
    else:
        print("Customer not found.")


def update_stage():
    name = input("Customer Name: ")

    if name in customers:
        print("1. Lead")
        print("2. Contacted")
        print("3. Proposal Sent")
        print("4. Won")
        print("5. Lost")

        choice = input("Select stage: ")

        stages = {
            "1": "Lead",
            "2": "Contacted",
            "3": "Proposal Sent",
            "4": "Won",
            "5": "Lost"
        }

        if choice in stages:
            customers[name].stage = stages[choice]
            print("Sales stage updated.")
        else:
            print("Invalid stage.")
    else:
        print("Customer not found.")


while True:
    print("\n===== CRM SYSTEM =====")
    print("1. Add Customer")
    print("2. View Customers")
    print("3. Add Communication Log")
    print("4. View Communication Logs")
    print("5. Update Sales Pipeline Stage")
    print("6. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_customer()

    elif choice == "2":
        view_customers()

    elif choice == "3":
        add_log()

    elif choice == "4":
        view_logs()

    elif choice == "5":
        update_stage()

    elif choice == "6":
        break

    else:
        print("Invalid option.")
```

```python
#online exam
class Question:
    def __init__(self, question, options, answer):
        self.question = question
        self.options = options
        self.answer = answer


class Student:
    def __init__(self, name):
        self.name = name
        self.score = 0


questions = []
students = []


def add_question():
    q = input("Enter question: ")
    options = []

    for i in range(4):
        opt = input(f"Option {i+1}: ")
        options.append(opt)

    answer = int(input("Correct option number (1-4): "))
    questions.append(Question(q, options, answer))

    print("Question added successfully.")


def register_student():
    name = input("Enter student name: ")
    students.append(Student(name))
    print("Student registered.")


def start_exam():
    name = input("Enter student name: ")

    student = None
    for s in students:
        if s.name == name:
            student = s
            break

    if not student:
        print("Student not registered.")
        return

    score = 0

    for q in questions:
        print("\n", q.question)

        for i, opt in enumerate(q.options):
            print(f"{i+1}. {opt}")

        ans = int(input("Your answer: "))

        if ans == q.answer:
            score += 1

    student.score = score
    print("Exam completed!")


def show_results():
    print("\n--- Results ---")

    for s in students:
        print(f"{s.name} : {s.score}/{len(questions)}")


while True:
    print("\n===== ONLINE EXAMINATION SYSTEM =====")
    print("1. Add Question")
    print("2. Register Student")
    print("3. Start Exam")
    print("4. Show Results")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_question()

    elif choice == "2":
        register_student()

    elif choice == "3":
        start_exam()

    elif choice == "4":
        show_results()

    elif choice == "5":
        break

    else:
        print("Invalid choice")
```
